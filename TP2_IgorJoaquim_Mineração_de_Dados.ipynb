## Metodologia

Neste projeto, seguimos a metodologia CRISP-DM (Cross-Industry Standard Process for Data Mining), que é amplamente reconhecida e adotada em projetos de mineração de dados. A metodologia CRISP-DM divide o processo de mineração de dados em seis fases inter-relacionadas, permitindo uma abordagem sistemática e estruturada:

### 1. Entendimento do Negócio
Na primeira fase, concentramos nossos esforços em compreender os objetivos de negócio, as necessidades da sociedade e as questões-chave que este projeto busca responder. Esta etapa envolve discussões com as partes interessadas para definir os objetivos específicos, bem como para estabelecer métricas de sucesso. A motivação principal é criar uma base sólida para a análise de dados, garantindo que os resultados sejam relevantes e úteis.

### 2. Entendimento dos Dados
Com os objetivos de negócios definidos, prosseguimos para a fase de entendimento dos dados. Nesta etapa, exploramos o conjunto de dados do ENEM 2022 para compreender sua estrutura, qualidade e conteúdo. Isso inclui a identificação de variáveis relevantes, tratamento de valores ausentes, e a seleção de dados relevantes para atender aos objetivos da mineração.

### 3. Preparação dos Dados
Uma vez que compreendemos os dados, iniciamos a fase de preparação dos dados. Isso envolve limpeza, transformação e formatação dos dados para que possam ser usados na modelagem. Também é o momento de criar novas features, quando necessário, e garantir que os dados estejam prontos para análise.

### 4. Modelagem
Na fase de modelagem, aplicamos técnicas de mineração de dados para atender aos objetivos da mineração. Isso inclui a criação de clusters com base em indicadores socioeconômicos, a análise de respostas dos participantes para identificar padrões, a seleção de métricas de similaridade e a avaliação do poder preditivo de variáveis relevantes. Além disso, desenvolvemos um classificador com base em indicadores socioeconômicos para atribuir alunos a clusters.

### 5. Avaliação
Após a modelagem, é fundamental avaliar a qualidade e eficácia dos resultados. Isso envolve a análise da qualidade dos clusters, a precisão do classificador, e a importância das variáveis preditivas. A avaliação nos ajuda a entender até que ponto os modelos criados atendem aos objetivos de negócios e permite ajustes, se necessário.

Esta metodologia proporciona um processo estruturado e iterativo para o projeto, garantindo que os resultados da mineração de dados sejam relevantes e aplicáveis aos objetivos de negócios e às necessidades da sociedade. Através das fases do CRISP-DM, estabelecemos uma abordagem sólida para a análise dos microdados do ENEM 2022, com o objetivo de promover a equidade na educação.

##**2. Data understanding (Entendimento dos dados)**
<p align="justify">O dataset "Amostra ENEM 2022" é uma coleção de 5,6 milhões participantes do Exame Nacional do Ensino Médio (Enem) de 2022. O dataset foi disponibilizado pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (INEP) e contém informações sobre as características socioeconômicas, educacionais e de desempenho dos participantes.

```python
#Carregando a base de dados
text_columns  = ['TX_RESPOSTAS_LC','TX_RESPOSTAS_CH','TX_RESPOSTAS_CN','TX_RESPOSTAS_MT']
grade_columns = ['NU_NOTA_LC'     ,'NU_NOTA_CH'     ,'NU_NOTA_CN'     ,'NU_NOTA_MT','NU_NOTA_REDACAO']
SES_columns   = [
                  "TP_SEXO",
                  "TP_ESTADO_CIVIL",
                  "TP_COR_RACA",
                  "TP_FAIXA_ETARIA",
                  "TP_NACIONALIDADE",
                  "TP_ST_CONCLUSAO",
                  "TP_ANO_CONCLUIU",
                  "TP_ESCOLA",
                  "IN_TREINEIRO",
                ]
op_columns    = [
                  'NU_INSCRICAO',
                  "CO_PROVA_CN",
                  "TP_PRESENCA_CN",
                  "TP_PRESENCA_CH",
                  "TP_PRESENCA_LC",
                  "TP_PRESENCA_MT",
                ]
question_columns = [f'Q{i:03}' for i in range(1, 26)]
```

```python
columns = text_columns + grade_columns + SES_columns + op_columns + question_columns

df = pd.read_csv("microdados.csv",sep=";",encoding="latin",usecols=columns)
print(df.shape)
```

O conjunto completo de dados possui 3.476.105 linhas  e 49 colunas

```python
df.head()
```

```python
isna    = df.isna().sum().values
dtypes  = df.dtypes.values
columns = list(set(df.columns) - set(question_columns))

max_col_width   = max(len(col) for col in columns)
max_nan_width   = max(len(str(isna_count)) for isna_count in isna)

for col, isna_count, dtype in zip(columns, isna, dtypes):
    print(f"{col:{max_col_width}} | NaNs: {isna_count:{max_nan_width}} | Type: {dtype}")
```

<p align="justify">Como podemos ver, os indicadores socioecômicos de todos os participantes são consistentes e não possuem dados faltantes, entretanto, houveram participantes que não prestaram todas as etapas da prova, resultando em notas, respostas e informações sobre provas faltantes.

<p align="justify"> Sendo assim, trataremos dados faltantes ao considerar apenas alunos que estiveram presentes em todas as etapas da prova que fizeram a prova laranja:

```python
# Define conditions for presence in each exam
condition_cn = (df['TP_PRESENCA_CN'] == 1) & (df['NU_NOTA_CN'] != 0)
condition_ch = (df['TP_PRESENCA_CH'] == 1) & (df['NU_NOTA_CH'] != 0)
condition_lc = (df['TP_PRESENCA_LC'] == 1) & (df['NU_NOTA_LC'] != 0)
condition_mt = (df['TP_PRESENCA_MT'] == 1) & (df['NU_NOTA_MT'] != 0)
condition_exam = df["CO_PROVA_CN"] == 1085  # prova azul

# Combine conditions to select rows with presence in all exams
available_participants = df[condition_cn & condition_ch & condition_lc & condition_mt & condition_exam].copy()
del df
```

Para tratar os dados faltantes e selecionar os alunos para a análise, foram estabelecidas condições criteriosas com base em múltiplos fatores. Inicialmente, as condições de presença em cada etapa do exame, representadas por `TP_PRESENCA_CN`, `TP_PRESENCA_CH`, `TP_PRESENCA_LC`, e `TP_PRESENCA_MT`, foram utilizadas para garantir a participação dos alunos em todas as avaliações. Além disso, as notas em cada uma das disciplinas, tais como `NU_NOTA_CN`, `NU_NOTA_CH`, `NU_NOTA_LC`, e `NU_NOTA_MT`, foram examinadas para eliminar notas igual a zero, garantindo que apenas notas válidas fossem consideradas. Finalmente, a seleção focou na prova laranja, identificada pelo código `CO_PROVA_CN` igual a 1085. O resultado desse processo é um conjunto de dados, denominado `available_participants`, composto apenas pelos alunos que atenderam a todas essas condições, proporcionando um conjunto completo e limpo para análise subsequente.

A função full_d_matrix e a matriz d_matrix fazem parte da etapa de preparação de dados. A função full_d_matrix é responsável por calcular a matriz de similaridade entre os participantes com base nas respostas corretas. Isso pode ser útil para criar features adicionais ou simplificar a representação dos dados. A matriz d_matrix é construída a partir das respostas corretas em diferentes áreas do ENEM.

```python
random_sample['RIGHT_ANSWERS_LC'] = random_sample["TX_RESPOSTAS_LC"].apply(lambda x: f(x,right_answers_LC))
random_sample['RIGHT_ANSWERS_CH'] = random_sample["TX_RESPOSTAS_CH"].apply(lambda x: f(x,right_answers_CH))
random_sample['RIGHT_ANSWERS_CN'] = random_sample["TX_RESPOSTAS_CN"].apply(lambda x: f(x,right_answers_CN))
random_sample['RIGHT_ANSWERS_MT'] = random_sample["TX_RESPOSTAS_MT"].apply(lambda x: f(x,right_answers_MT))
```

```python
# INÍCIO CODIGO MANUAL
def full_d_matrix(series):
  cupy_matrix   = cp.vstack(series.apply(cp.array).values)
  intersection  = cp.logical_and(cupy_matrix[:, None], cupy_matrix)
  intersection  = cp.sum(intersection, axis=2)
  union  = cp.logical_or(cupy_matrix[:, None], cupy_matrix)
  union  = cp.sum(union, axis=2)

  return cp.divide(intersection,union)

d_matrix  = full_d_matrix(random_sample["RIGHT_ANSWERS_MT"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_CH"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_CN"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_LC"])
# FIM CODIGO MANUAL
```

```python
d_matrix /=4
```

```python
d_matrix = cp.asnumpy(d_matrix)
d_matrix = np.nan_to_num(d_matrix, nan=0)
sns.histplot(d_matrix.reshape(-1))
plt.show()
```

Aqui é onde a etapa de modelagem é executada. A função get_spectral_clustering aplica o algoritmo de clusterização espectral à matriz de similaridade d_matrix. Essa é uma técnica de aprendizado não supervisionado que visa agrupar participantes com base em sua similaridade nas respostas corretas.

```python
def get_spectral_clustering(d_matrix):
  n_clusters = 3
  spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors',random_state=0)
  labels = spectral_clustering.fit_predict(d_matrix)
  return labels

def show_grades_by_cluster(df,labelss):
  # Group the data by cluster labels and calculate the centroids
  cluster_centers = df.groupby(labelss)[grade_columns].mean()

  mean_of_cluster_centers = np.mean(cluster_centers, axis=1)
  cluster_centers = cluster_centers.assign(Mean_Centroids=mean_of_cluster_centers)
  labels = pd.qcut(mean_of_cluster_centers, q=3, labels=["Low", "Medium", "High"])

  df[labelss] = df[labelss].map(labels)
```

```python
labels = get_spectral_clustering(d_matrix)
random_sample["Graph_Labels"] = labels
show_grades_by_cluster(random_sample,"Graph_Labels")
```

##**6. Melhora do Modelo**


A justificativa para realizar a comparação entre clusters utilizando o código apresentado é avaliar e selecionar o número ótimo de clusters (k) para o nosso conjunto de dados. Esse processo é fundamental em análise de cluster, pois nos ajuda a determinar o agrupamento mais apropriado para representar a estrutura subjacente dos dados.

```python
d_matrix  = full_d_matrix(random_sample["RIGHT_ANSWERS_MT"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_CH"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_CN"])
d_matrix += full_d_matrix(random_sample["RIGHT_ANSWERS_LC"])
```

```python
X = cp.asnumpy(d_matrix)/4
X = np.nan_to_num(X, nan=0)
np.fill_diagonal(X, 0)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit and transform the data using the scaler
standardized_data = scaler.fit_transform(random_sample[grade_columns])
```

```python
distance_matrix = pairwise_distances(random_sample[grade_columns], metric='euclidean')
np.fill_diagonal(distance_matrix, 0)
```

O Silhouette Score é uma métrica comumente usada para avaliar a qualidade dos clusters gerados. Ele mede o quão similar um objeto é ao seu próprio cluster (coesão) em comparação com outros clusters (separação). Um Silhouette Score mais alto indica clusters mais bem definidos e distintos.

O Silhouette Score pode ser usado para determinar o número ótimo de clusters. Quanto maior o valor do Silhouette Score, melhor. Portanto, ao calcular o Silhouette Score para diferentes valores de k, podemos identificar o valor de k que resulta nos clusters mais bem definidos.

O método do Elbow ajuda a evitar problemas de overfitting (muitos clusters pequenos que capturam o ruído) e underfitting (poucos clusters grandes que perdem a estrutura dos dados). Encontrar o "Elbow" fornece um equilíbrio entre complexidade do modelo e capacidade de representar os dados.

```python
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.cluster._spectral")
warnings.filterwarnings("ignore", category=FutureWarning, module="sklearn.cluster._kmeans")
k_values = range(2, 11)  # Adjust the range as needed

# Lists to store metrics
silhouette_scores_spectral = []
explained_variance_spectral = []
silhouette_scores_kmeans = []
explained_variance_kmeans = []

# Perform clustering and calculate metrics for each k
for k in k_values:
    # Spectral Clustering
    spectral_clustering = SpectralClustering(n_clusters=k, affinity='nearest_neighbors', random_state=0)
    labels_spectral     = spectral_clustering.fit_predict(X)
    silhouette_spectral = silhouette_score(distance_matrix, labels_spectral, metric='precomputed')
    silhouette_scores_spectral.append(silhouette_spectral)
    explained_variance_spectral.append(np.var(labels_spectral))

    # K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels_kmeans = kmeans.fit_predict(standardized_data)
    silhouette_kmeans = silhouette_score(distance_matrix, labels_kmeans, metric='precomputed')
    silhouette_scores_kmeans.append(silhouette_kmeans)
    explained_variance_kmeans.append(np.var(labels_kmeans))
```

```python
# Plot Silhouette scores and explained variance for both methods
plt.figure(figsize=(12, 6))

# Silhouette scores
plt.subplot(1, 2, 1)
plt.plot(k_values, silhouette_scores_spectral, marker='o', label='Spectral Clustering', linestyle='-', color='b')
plt.plot(k_values, silhouette_scores_kmeans, marker='o', label='K-means', linestyle='-', color='r')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)
plt.legend()

# Explained variance (Elbow method)
plt.subplot(1, 2, 2)
plt.plot(k_values, explained_variance_spectral, marker='o', label='Spectral Clustering', linestyle='-', color='b')
plt.plot(k_values, explained_variance_kmeans, marker='o', label='K-means', linestyle='-', color='r')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Explained Variance')
plt.title('Explained Variance vs. Number of Clusters (Elbow Method)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

A escolha do método de linkage 'ward' é comum no clustering hierárquico, pois ele visa minimizar a variância dentro dos clusters, o que muitas vezes resulta em clusters mais bem definidos. Além disso, o código foi recomendado porque você indicou que desejava usar um distance matrix (matriz de distância) para realizar o clustering.

```python
Z = linkage(X, method='ward')

k_values = range(2, 11)  # Adjust the range as needed

silhouette_scores = []

# Perform hierarchical clustering for each k
for k in k_values:
    labels = fcluster(Z, k, criterion='maxclust')

    silhouette = silhouette_score(distance_matrix, labels, metric='precomputed')
    silhouette_scores.append(silhouette)

# Plot Silhouette scores for different k values
plt.figure(figsize=(10, 5))
plt.plot(k_values, silhouette_scores, marker='o', linestyle='-', color='b')
plt.plot(k_values, silhouette_scores_kmeans, marker='o', label='K-means', linestyle='-', color='r')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters (Hierarchical Clustering)')
plt.grid(True)

plt.show()
```
